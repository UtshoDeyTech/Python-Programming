```python
from collections import deque

def bfs(graph, start, item):
    queue = deque([start])
    visited = set()
    while queue:
        node = queue.popleft()
        if node == item:
            return True
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                queue.append(neighbor)
    return False
```

The bfs function takes a graph, a starting node, and an item to search for as its parameters. It initializes a queue with the starting node and an empty visited set. Then, it enters a loop that continues until the queue is empty. In each iteration of the loop, it removes the first node from the queue and checks if it is the search item. If it is, it returns True. Otherwise, it adds the node to the visited set and adds all of its unvisited neighbors to the queue. If the item is not found after visiting all reachable nodes, the function returns False.



```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}
start_node = 'A'
search_item = 'F'
found = bfs(graph, start_node, search_item)
if found:
    print(f"{search_item} is found in the graph.")
else:
    print(f"{search_item} is not found in the graph.")
```
