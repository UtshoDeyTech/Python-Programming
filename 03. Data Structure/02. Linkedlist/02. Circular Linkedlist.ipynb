```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next_node = None

```

```python
class CircularLinkedList:
    def __init__(self):
        self.head = None

    # Add item to end of list
    def add_item(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next_node = self.head
        else:
            current_node = self.head
            while current_node.next_node != self.head:
                current_node = current_node.next_node
            current_node.next_node = new_node
            new_node.next_node = self.head

    # Remove item from list
    def remove_item(self, data):
        if not self.head:
            return
        elif self.head.data == data:
            current_node = self.head
            while current_node.next_node != self.head:
                current_node = current_node.next_node
            current_node.next_node = self.head.next_node
            self.head = self.head.next_node
        else:
            current_node = self.head
            previous_node = None
            while current_node.next_node != self.head:
                previous_node = current_node
                current_node = current_node.next_node
                if current_node.data == data:
                    previous_node.next_node = current_node.next_node
                    break

    # Return index of item in list
    def item_index(self, data):
        current_node = self.head
        index = 0
        while current_node.next_node != self.head:
            if current_node.data == data:
                return index
            current_node = current_node.next_node
            index += 1
        if current_node.data == data:
            return index
        return -1

    # Return size of list
    def size(self):
        if not self.head:
            return 0
        current_node = self.head
        count = 1
        while current_node.next_node != self.head:
            count += 1
            current_node = current_node.next_node
        return count

    # Return whether list is empty
    def is_empty(self):
        return not self.head

    # Clear list
    def clear(self):
        self.head = None

    # Return string representation of list
    def __str__(self):
        if not self.head:
            return ''
        current_node = self.head
        str_list = []
        while current_node.next_node != self.head:
            str_list.append(str(current_node.data))
            current_node = current_node.next_node
        str_list.append(str(current_node.data))
        return ' -> '.join(str_list)
```

```python
# Create circular linked list object
my_list = CircularLinkedList()

# Add items
my_list.add_item('a')
my_list.add_item('b')
my_list.add_item('c')
my_list.add_item('d')

# Print list
print(my_list)  # Output: a -> b -> c -> d

# Remove item
my_list.remove_item('c')

# Print list
print(my_list)  # Output: a -> b -> d

# Get index of item
print(my_list.item_index('b'))  # Output: 1

# Get size of list
print(my_list.size())  # Output: 3

# Check if list is empty
print(my_list.is_empty())  # Output: False

# Clear list
my_list.clear()

# Check if list is empty
print(my_list.is_empty())  # Output: True
```
